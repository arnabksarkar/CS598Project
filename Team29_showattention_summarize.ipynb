```python
#import pip
#pip.main(["install", "openpyxl"])
import csv
import pandas as pd
import numpy  as np
import math
#import openpyxl
#import xlrd

target = 'VANCOMYCIN'
time_steps = 14
person_count = 0
features_count = 0
activation_order_threshhold = 10 #top number of activations to count
```

```python
dfs = pd.read_excel('./saved_csvs/activations_order_{0}.xlsx'.format(target))
print(dfs.head)
```

```python
#Get the columns and make a matrix as follows
#column name, count of activations at or below a certain level
header_list = list(dfs)
print(header_list)
feature_list = []
for i in range(len(header_list) - 4):
    feature_list.append(header_list[i+4])
#print(len(feature_list), feature_list)  
feature_count = len(feature_list)

feature_howmany = np.zeros(feature_count, dtype=int)
print('{0} feature_howmany'.format(target), feature_howmany)
#for irow in range(dfs.shape[0] - 1):
#for index, row in dfs.iloc[:10].iterrows():
for index, row in dfs.iterrows():
    row_list = list(row)
    #print("row_list:", row_list)
    
    if isinstance(row_list[3],int) and row_list[3] >= 0:
        #print("index, row_list:", index, row_list)
        for icol in range(4, len(row_list)):
            #print("icol:", icol, " = ", row_list[icol])
            if row_list[icol] <= activation_order_threshhold and row_list[icol] >= 0:
                feature_howmany[icol - 4] += 1
                #print("icol:", icol, " = ", row_list[icol], " has threshhold ge 0 and le ", activation_order_threshhold)

#zip the lists and write them to a csv file
f = open('./saved_csvs/activations_counts_{0}_threshhold_{1}.csv'.format(target, activation_order_threshhold), 'w', newline='') #open output file
writer = csv.writer(f) #create the csv writer
print(target, "feature_list", "feature_howmany")
for i in range(feature_count):
    print(feature_list[i],",",feature_howmany[i])

line_list = []
for i in range(features_count):
    line_list = feature_list[i] + "," + feature_howmany[i] 
    print(line_list)
    writer.writerow(line_list)
    
f.close()
```
